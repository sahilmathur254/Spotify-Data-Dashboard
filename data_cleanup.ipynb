```python
import pandas as pd
import numpy as np
```

```python
#importing data from excel
df_song_list = pd.read_excel('Most Streamed Spotify Songs 2024.xlsx', sheet_name='Spotify Song List')
df_stream_detail = pd.read_excel('Most Streamed Spotify Songs 2024.xlsx', sheet_name='Spotify Stream Detail')
```

```python
df_song_list.head()
```

```python
df_stream_detail.head()
```

```python
print(df_stream_detail.shape)
print(df_song_list.shape)
```

```python
#concatenate the two dataframes based on ID
df = pd.merge(df_song_list, df_stream_detail , on='ID')
df.head()
```

```python
df.duplicated().sum()
```

```python
df[df['Album Name'].isnull()]
```

```python
df[df['Artist'].isnull()]
```

```python
#filling the null values with 'singles' in 'Album Name' Column
df['Album Name'] = df['Album Name'].fillna('Singles')
#filling the null values with 'Unknown Artists' in 'Artist' Column
df['Artist'] = df['Artist'].fillna('Unknown Artist')

#Dropping ISRC and 'TIDAL Popularity' columns
df = df.drop(columns=['ISRC', 'TIDAL Popularity'])
```

```python
df.isnull().sum()
```

```python
#fill the null values with 0 for all the columns
df = df.fillna(0)
```

```python
#converting all the dates in 'Release Date' to date format
df['Release Date'] = pd.to_datetime(df['Release Date']).dt.strftime('%m/%d/%Y')
```

```python
df.head()
```

```python
#export the cleaned data to a new excel file
df.to_excel('Most Streamed Spotify Songs 2024_cleaned.xlsx', index=False)
```
